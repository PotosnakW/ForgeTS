```python
import pickle
import numpy as np

with open("/home/wpotosna/forgets/results/results_m1_yearly_arima.pkl", "rb") as f:
    res = pickle.load(f)

print(res["dataset"], f"W={res['W']} H={res['H']} step={res['step']}")
print(f"{len(res['summary'])} entries")

# aggregate: does lag-norm beat causal-norm (or vice versa)?
mae_lag  = np.concatenate([s["mae_lag"]  for s in res["summary"]])
mae_caus = np.concatenate([s["mae_caus"] for s in res["summary"]])
print(f"MAE lag-norm:    {mae_lag.mean():.4f} ± {mae_lag.std():.4f}")
print(f"MAE causal-norm: {mae_caus.mean():.4f} ± {mae_caus.std():.4f}")
print(f"lag wins: {(mae_lag < mae_caus).mean():.1%} of origins")
```
